## Importing the required libraries

**pandas :** pandas is a Python package that provides fast, flexible, and expressive data structures designed to make working with structured (tabular, multidimensional, potentially heterogeneous) and time series data both easy and intuitive.

---

**Matplotlib :** Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is a multi-platform data visualization library built on NumPy arrays and designed to work with the broader SciPy stack.

```
import pandas as pd
import matplotlib.pyplot as plt
```

```
#path of the dataset
path = '/content/stock_data2.csv' 
```

```
#Loading the dataset as a DataFrame using pandas pd.DataFrame() functiom
df = pd.read_csv(path)
```

## Data Visualizations

`df.plot()` - Make plots of Series or DataFrame.
default - Line plot/ graph.

### Plotting a single line

```
#Plotting Stock Prices of Company-1 vs Days
df.plot('day', 'company1')
plt.show()
```

```
#Plotting Stock Prices of Company-2 vs Days
df.plot('day', 'company2')
plt.show()
```

### Adding two lines on the same plot

```
plt.plot(df['day'], df['company1']) #Stock Prices of Company-1
plt.plot(df['day'], df['company2']) #Stock Prices of Company-2
plt.show()
```

### Adding labels to the lines

```
plt.plot(df['day'], df['company1'], label='Company-1') #Adding labels
plt.plot(df['day'], df['company2'], label='Company-2') #Adding labels
plt.legend()
plt.show()
```

### Different Linestyles and colors

```
plt.plot(df['day'], df['company1'], label='Company-1', linestyle='--', color='red') #Adding different linestyle and color to the line
plt.plot(df['day'], df['company2'], label='Company-2', linestyle='-', color='green') #Adding different linestyle and color to the line
plt.legend()
plt.show()
```

### Adding axis labels and title

```
plt.plot(df['day'], df['company1'], label='Company-1', linestyle='--', color='red')
plt.plot(df['day'], df['company2'], label='Company-2', linestyle='-', color='green')
plt.xlabel('Days') #Adding x-label
plt.ylabel('Prices, $') #Adding y-label
plt.title('Company Stock Prices over Time') #Adding title to the plot
plt.legend()
plt.show()
```

```
#Computing mean of stock prices of Company-1
df['company1_mean'] = df['company1'].mean()
```

```
#Computing mean of stock prices of Company-2
df['company2_mean'] = df['company2'].mean()
```

### Additional Lines to the plot

```
plt.plot(df['day'], df['company1'], label='Stock Prices')
plt.plot(df['day'], df['company1_mean'], label='Mean Price')
plt.legend()
plt.xlabel('Days')
plt.ylabel('Prices, $')
plt.title('Company-1 Stock Prices over Time ')
plt.show()
```

```
plt.plot(df['day'], df['company2'], label='Stock Prices', color='purple')
plt.plot(df['day'], df['company2_mean'], label='Mean Price', color='cyan')
plt.legend()
plt.xlabel('Days')
plt.ylabel('Prices, $')
plt.title('Company-2 Stock Prices over Time ')
plt.show()
```

### Layering Histograms on a single plot

```
plt.hist(df['company1'], bins=100, label='Company-1')
plt.hist(df['company2'], bins=100, label='Company-2')
plt.legend()
plt.show()
```

### Alpha - Changing transparency of histograms

```
plt.hist(df['company1'], bins=100, label='Company-1', alpha=0.4)
plt.hist(df['company2'], bins=100, label='Company-2', alpha=0.4)
plt.legend()
plt.xlabel('Stock Prices')
plt.show()
```
